```python
# ----- Task 1: Data Parsing and Profile Cleaning -----
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

# Cleaning and validating the data for student in raw_students
for student in raw_students:
    
    # Name Cleaning 
    name = student["name"]
    name = name.strip()
    name = name.title()
    
    # Roll Number Cleaning 
    roll = student["roll"]
    roll = roll.strip()
    roll = int(roll)
    
    # Marks Cleaning 
    mark_str = student["marks_str"]
    mark_str = mark_str.split(",")
    mark_list = []
    
    # loop through each mark string
    for mark in mark_str:
        mark = mark.strip()
        mark = int(mark)
        mark_list.append(mark)
        
    # Name Validation 
    temp_name = name.replace(" ", "")
    
    # Verify that name contains only alphabetic charaacters
    if temp_name.isalpha():
        print("Student Name : ", name, " - Valid name")
    else:
        print("Student Name : ", name, " - Invalid name")
        
    # Display cleaaned roll number and marks
    print("Roll number : ", roll)
    print("Marks :", mark_list)
    print("----------------------------------------")
    
    # ----- SPECIAL CONDITION CHECK -----
    # Additional formatting for student with roll number 103
    if roll == 103:
        print("Checking Roll number 103 for special formatting...")
        print("Formatted Name : ", name.upper())
        print("Formatted Name : ", name.lower())
        print("----------------------------------------")
```

```python
# ----- Task 2: Marks Analysis Loops and Conditional Statements -----
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]

# ----- Loop through marks and display subject-wise marks ----- 
print("Report for : ", student_name)

# Subject-wise Grades 
for i in range(len(subjects)):
    sub = subjects[i]
    mark = marks[i]
    
    # Grade calculation based on marks
    if mark >= 90:
        grade = "A+"
    elif mark >= 80:
        grade = "A"
    elif mark >= 70:
        grade = "B"
    elif mark >= 60:
        grade = "C"
    else:
        grade = "F"
    
    print(sub, " : ", mark, " - Grade : ", grade)
    
# ----- Summary of marks -----

# Calculate total and average marks 
total_marks = sum(marks)
average_marks = total_marks / len(marks)

# Calculate the highest and lowest marks 
highest_mark = max(marks)
highest_subject = subjects[marks.index(highest_mark)]

lowest_mark = min(marks)
lowest_subject = subjects[marks.index(lowest_mark)]

# ----- Display the summary -----
print("-----------------------------------------")
print("Total marks : ", total_marks)
print("Average marks :", format(average_marks, ".2f"))
print("Highest Subject : ", highest_subject, "(", highest_mark, ")")
print("Lowest Subject : ", lowest_subject, "(", lowest_mark, ")")
print("-----------------------------------------")

# ----- Input for new subject mrks -----
print("Enter new subject (type done to finish)")

count = 0

while True:
    sub = input("Enter Subject : ")
    
    if sub.lower() == "done" :
        break
    
    mark_input = input("Enter marks : ")
    
    if mark_input.isdigit():
        mark = int(mark_input)
        
        if 0 <= mark <= 100:
            subjects.append(sub)
            marks.append(mark)
            count += 1
            print("Subject added successfully!")
        else:
            print("Invalid marks. Please enter a value between 0 and 100.")
    else:
        print("Invalid input")
        
# ----- Updated Result ----- 
# New total and average 
new_total = sum(marks)
new_avg = new_total/len(marks)

print("-----------------------------------------")
print("Updated Total marks : ", new_total)
print("Updated Average marks :", format(new_avg, ".2f"))
print("-----------------------------------------")
```

```python
# ----- Task 3: Class Performance Summary -----
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# ----- Counter and Tracking Variables -----
total_students = len(class_data)

pass_count = 0
fail_count = 0
avg_list = []

# ----- Top name and average -----
top_name = ""
top_avg = 0

# ----- Display the list -----
print("Class Performance Summary")
print("Name  |  Average Marks  |  Status")
print("-----------------------------------------")

# ----- Loop through each student -----
for student in class_data:
    name = student[0]
    marks = student[1]
    
    total = sum(marks)
    avg = total/len(marks)
    avg_list.append((avg, format(avg, ".2f")))
    
    if avg >= 60:
        status = "Pass"
        pass_count +=1
        
    else:
        status = "Fail"
        fail_count += 1
        
    if avg > top_avg:
        top_avg = avg
        top_name = name
        
    print(name, " | ", format(avg, ".2f"), " | ", status)
    
# ----- Class Summary -----
class_avg = sum(avg for avg, _ in avg_list) / len(avg_list)
class_avg = float(format(class_avg, ".2f"))

print("-----------------------------------------")
print("Total Students : ", total_students)
print("Total Pass : ", pass_count)
print("Total Fail : ", fail_count)
print("Class Average : ", format(class_avg, ".2f"))
print("Top Student : ", top_name, "(", format(top_avg, ".2f"), ")")
    
```
